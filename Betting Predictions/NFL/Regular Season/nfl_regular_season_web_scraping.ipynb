```python
from selenium import webdriver
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
from selenium.webdriver.edge.service import Service

# Set up the Selenium WebDriver (make sure to have the appropriate webdriver installed)
service = Service(executable_path="C:\Program Files (x86)\msedgedriver.exe")
driver = webdriver.Edge(service=service)

# URL of the webpage containing the dropdown menu
url = "https://www.footballdb.com/statistics/nfl/team-stats/offense-totals/2025/regular-season?sort=totpts"
driver.get(url)
```

```python

# REGULAR SEASON PREDICTIONS #
# Click on dropdown to get years
drop_down_menu_year = driver.find_element(By.XPATH, '//*[@id="dropdownMenuYear"]').click()

# Get years
years = driver.find_element(By.XPATH,'//*[@id="leftcol"]/div[2]/div[1]/div[2]/ul').text.split('\n')[1:]

# Get data for each year
table_data = []
for year in years:
    # Visit the site for every year
    url = f"https://www.footballdb.com/statistics/nfl/team-stats/offense-totals/{year}/regular-season?sort=totpts"
    driver.get(url)

    # Find the table on the page
    table = WebDriverWait(driver, 2).until(
    EC.presence_of_element_located((By.XPATH, '//*[@id="leftcol"]/div[3]/div[2]/div[2]/table'))
    )

    # Store the table data in a variable for looping
    tbody = table.find_element(By.TAG_NAME, 'tbody')
    rows = WebDriverWait(driver, 2).until(
        EC.presence_of_all_elements_located((By.TAG_NAME, 'tr'))
    )

    for row in rows:
        try:
            row_cells = row.find_elements(By.TAG_NAME, 'td')
            # Check if the row contains cells (data) before processing
            if row_cells:
                row_text = [cell.text for cell in row_cells]
                row_text.append(year)
                table_data.append(row_text) 
                print(f"Adding {year} data for {row_text[0]}")
        except:
            # Handle StaleElementReferenceException by refinding the row
            rows = WebDriverWait(driver, 2).until(
                EC.presence_of_all_elements_located((By.TAG_NAME, 'tr'))
            )
# Get table headers
headers = table.find_element(By.XPATH,'//*[@id="leftcol"]/div[3]/div[2]/div[2]/table/thead').text.split('\n')
headers.append('Season')

# Close the browser window
driver.quit()  
```

```python
import pandas as pd

# Getting the headers from the data array
# It is important to remove them from the data array afterwards in order to be parsed correctly by Pandas
team_stats = pd.DataFrame(table_data, columns=headers)
team_stats
```

# Cleaning

```python
# Check data types
team_stats.info()
```

```python
team_stats.columns
```

```python
# Columns in thousands have ('#,###)
columns_in_thous = ['RushYds', 'PassYds', 'TotYds']

# Remove comma ('2,875' -> '2875')
for col in columns_in_thous:
    team_stats[col] = team_stats[col].str.replace(',', '')
```

```python
# Convert following columns to floats
string_columns = ['Gms', 'Pts', 'Pts/G', 'RushYds', 'RYds/G', 'PassYds', 'PYds/G','TotYds', 'Yds/G']
team_stats[string_columns] = team_stats[string_columns].apply(pd.to_numeric, errors='coerce')

team_stats.info()
```

```python
file_path = 'nfl_regular_season_team_stats.csv'
team_stats.to_csv(file_path, index=False)
```
